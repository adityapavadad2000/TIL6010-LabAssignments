## Question 1:

Recreate the barplot below that shows the population of different continents for the year 2007.

## Question 4:

Thus far we looked at data from one year (2007). Lets create an animation to see the population growth of the continents through the years


```python
import plotly.express as px

fig = px.histogram(df, 
             x='continent', 
             y='pop',
             animation_frame='year',  
             title='Population Growth by Continent Over the Years',
             labels={'pop': 'Population'},
             category_orders={"continent": df.groupby("continent")["pop"].sum().sort_values(ascending=False).index},
             height=600,  
             color = 'continent'
            )

fig.update_layout(xaxis_title='Continent', yaxis_title='Population',  title_x=0.5, title_y=0.95)
# fig.update_xaxes(type='category')

fig.show()
```

## Question 5:

Instead of the continents, lets look at individual countries. Create an animation that shows the population growth of the countries through the years

```python
# YOUR CODE HERE
fig = px.bar(df, 
             x='pop', 
             y='country',
             animation_frame='year', 
             title='Population Growth by Continent',
             labels={'pop': 'Population'},
             category_orders={"country": df.groupby("country")["pop"].sum().sort_values(ascending=False).index},
             height=600, 
             color = 'continent'
            )


fig.update_layout(xaxis_title='Continent', yaxis_title='Population' , title_x=0.5, title_y=0.95)
# fig.update_xaxes(type='category')

fig.show()
```

## Question 7:

Show only the top 10 countries in the animation

Hint: Use the axis limit to set this. 

```python

top_10 = df.groupby('year').apply(lambda x: x.nlargest(10, 'pop')).reset_index(drop=True)


fig = px.bar(top_10, 
             x='country', 
             y='pop',
             animation_frame='year',
             title='Top 10 Countries by Population Over the Years',
             height=600,
             color = "country"
            )


fig.update_layout(xaxis_title='Country', yaxis_title='Population' ,title_x=0.5, title_y=0.95)
fig.update_xaxes(type='category')
fig.update_xaxes( categoryorder = 'total descending')

fig.show()
```
